#### b- ¿Cómo y con qué comandos guardaría la mayor cantidad de detalle sobre las salidas de un script python que desea ejecutar de forma diaria a las 6AM?

Para utilizar `cron` se debe especificar un lista de trabajos via `crontab` (cron table) con el siguiente formato

<img src="./img/cron.png" />

Comandos:

\# Modificamos la lista de trabajos con un editor en modo edicion `-e`<br/>
`env EDITOR=[someEditor] crontab -e`

\# Agregamos el trabajo que queremos repetir en forma diaria a las 6AM <br/>
`0 6 * * *  cd ~/path/to/script && python3 script.py >> script.out` 


Notas:
- Notar como podemos especificar el path hasta el directorio que contiene el script y usar `&&` para encadenar comandos <br/>
- En caso de no reconocer `python3` podemos especificar su path `/usr/bin/python3` (obtener con `whereis python3`)
- Podemos guardar el output de nuestro script en `script.out` utilizando `>>` para tener mas detalles del output

#### d- Si una instancia de Redshift utilizada para reporting se está quedando sin espacio y se impone la necesidad de sacar algunos datos antiguos de la base, pero a pesar de que los datos de más de seis meses de antigüedad no se utilicen para reporting, se los requiere para entrenar y validar modelos predictivos, además de hacer algunos análisis ad-hoc en SQL <span style="color:red"> a un precio razonable considerando tanto infraestructura como costos de consultas </span> ¿Que tipo de solución propondría para poder consultar los datos usando servicios cloud en AWS?

TL-DR: Podriamos migrar los datos antiguos desde Redshift a un Data Lake

Podemos basar esta decision comparando nuestras opciones:

a) Data Warehouse: base de datos optimizada para el analisis de datos. La estructura y esquema de los datos esta bien definida para optimizar SQL queries complejos cuyos resultados son usados para reporte operacional y analisis. La data en warehouse debe encontrarse limpia, enriquecida y trasformada para optimizar su rapido analisis. Por consiguiente, el costo de almacenamiento de datos es mayor aqui. 
Usuarios tipicos: business analysts.
Tareas tipicas: Batch reporting, BI and visualizations

b) Data Lake: pueden ser utilizados para almacenar datos de fuentes muy variadas (i.e data relacional de operaciones de negocio, data no relacional de aplicaciones mobiles, social media, etc. Su estructura y eschema no esta definida como el caso de Warehouse (se puede definir al momento de su uso), por lo cual no se require tanto planeamiento y diseno para guardar los datos. Diferentes tipos de tecnicas como SQL queries, big data analytics y machine learning models pueden servirce de data en un Data Lake. En contraposicion a los Warehouses, el costo de almacenamiento en Lakes es menor. 

Usuarios tipicos: Data Scientists, Data Developers y Business Analysts (usando datos curados)
Tareas tipicas: Machine Learning, Predictive analytics, data discovery and profiling

Source: https://aws.amazon.com/big-data/datalakes-and-analytics/what-is-a-data-lake/
